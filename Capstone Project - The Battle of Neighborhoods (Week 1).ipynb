Dataset has 533crows and 22 columns.

To make this analysis we can use cluster analysis.

Another very useful analysis could be to compare these data with data from another city to calculate the incidence rate of crimes that occurred in both places and compare them. If a place is more affected by crime, the state could act by increasing the police in that city and trying to understand what is due to the increased crime.
